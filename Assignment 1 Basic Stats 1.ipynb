## Standardization of numerical columns

```python
# Steps:-
# 1.Explain standardization concept
# 2.Standardize numerical columns
```

```python
# Standardization :-
  # The dataset we use have different variables which are measured with help of different units.
  # Secondly, sometimes data might be very big, and create mess while building an algorithm. 
  # So to make this data much more precise and maintain relationship between different variables we use standardization process
  # Standardization is process where data is centered around mean with std deviation of 1.
  # It helps to transform the values of variables in a data set to a similar scale.
  # This process is less sensitive to outliers.
  # It changes shape of data but maintains relation between multiple variables
```

```python
df.head()
```

```python
num_zscore = stats.zscore(df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']])
```

## Converting Categorical data in dummy variables

```python
# Need :-
    # Dataset contains multiple variables, some of which are categorical in nature.
    # With help of standardization, numerical variables can be brought to scale and used in our analysis 
    # But, it is hard to make use of categorical data in our analysis, which is why these variables are converted 
    # Dummy variables represent nothing but presence or absence of that particular variable or parameter
    # It is done by assigning 1 or 0 to that particular variable, 1 being present and 0 being absence of variable
    # One Hot Encoding is common process used to convert categorical data to dummy variables
```

```python
df.head()
```
