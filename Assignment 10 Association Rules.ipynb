```python
df = pd.read_csv('https://raw.githubusercontent.com/PrathameshR10/Assignments/main/Folder/Online%20retail.csv',header=None)
# pd.set_option('display.max_rows',None)
```

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
data = df[0].apply(txt_split)
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
len(te.columns_)
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
scores = apriori(data,min_support=0.03, use_colnames=True)
scores
```

```python
rules = association_rules(scores, metric="confidence", min_threshold=0.2)
rules
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
